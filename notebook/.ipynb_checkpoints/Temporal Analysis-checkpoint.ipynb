```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset_name = "sx-mathoverflow"
tau = 44 # number of snapshots
window_size = 2 # 
h = -1 # union cardinality
```

```python
degrees = pd.read_csv("../" + dataset_name + "_" + str(tau) + "_" + str(window_size) + "_" + str(h) + "_deg.csv")
degrees = degrees.loc[~(degrees.iloc[:,1:]==0).all(axis=1)]
degrees.reset_index(inplace=True, drop=True)
degrees
```

```python
coreness = pd.read_csv("../" + dataset_name + "_" + str(tau) + "_" + str(window_size) + "_" + str(h) + "_coreness.csv")
coreness = coreness.loc[~(coreness.iloc[:,1:]==0).all(axis=1)]
coreness.reset_index(inplace=True, drop=True)
coreness
```

```python
concatenated = pd.concat([coreness, degrees], axis=1)
concatenated
```

```python
for i in range(1,42):
    sns.set(style="whitegrid")
    sns.set_context("notebook", font_scale=1.2, rc={"lines.linewidth": 2.5})
    plt.yscale("log")
    plt.xscale("log")

    sns.scatterplot(concatenated, x="d_"+str(i), y="c_"+str(i))
    #sns.scatterplot(degrees, x=degrees.index, y=str(i), label="degree")
    plt.plot(np.arange(1, concatenated["d_"+str(i)].values.max()), np.arange(1, concatenated["d_"+str(i)].values.max()), color="orange")
    plt.title("Window # " + str(i))
    #plt.legend()
    plt.show()
```
